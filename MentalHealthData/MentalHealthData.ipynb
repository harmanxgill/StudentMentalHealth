## Create the Lexicon

```python
mental_health_lex = ['mental health', 'stress', 'anxiety', 'autism', 'bipolar', 'disorder', 'depression', 'ptsd', 'abuse', 'mental', 'substance abuse', 'alcoholism', 'behavioral', 'hysteria', 'madness', 'phobia', 'schizophrenia', 'therapy', 'trauma', 'self image', 'adhd']
```

## Mental Health Dataset

```python
post_ids = list(mentalhealth80_df['id'])
```

```python
print(post_ids)
```

```python
mentalhealth_df = pd.DataFrame(columns = ['id', 'author',	'author_flair_text', 'created_utc',	'num_comments',	'post', 'preprocessed_post'])
```

```python
for i in range(len(reddit_df)):
  item = reddit_df.loc[i, 'post'].lower().split()
  if reddit_df.loc[i, 'id'] in post_ids:
    mentalhealth_df = mentalhealth_df.append(reddit_df.iloc[i])
  else:
     for x in mental_health_lex:
       if x in item:
         mentalhealth_df = mentalhealth_df.append(reddit_df.iloc[i])
         break
```

```python
mentalhealth_df = mentalhealth_df.reset_index(drop=True)
```

```python
print(len(mentalhealth_df))
mentalhealth_df[:10]
```

```python
mentalhealth_df.to_json("MentalHealthDataset.json", orient="columns")
```

## Comments

```python
submission_ids = list(mentalhealth_df.loc[:, 'id'])
```

```python
print(submission_ids)
```

```python
api = PushshiftAPI()
```

```python
comments_ids = []
```

```python
print(type(comments_ids))
```

```python
%%time
for i in range(len(submission_ids)):
  submission_comments = api.search_submission_comment_ids(ids=submission_ids[i])
  ids = [_id for _id in submission_comments]
  comments_ids.append(ids)
```

```python
print(comments_ids)
```

```python
pd.DataFrame(comments_ids).to_csv("Comments.csv",index=False)
```
